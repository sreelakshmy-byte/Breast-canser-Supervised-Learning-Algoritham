```python
# Function to remove outliers using IQR
def remove_outliers_iqr(dataframe):
    df_cleaned = dataframe.copy()
    for column in df_cleaned.columns[:-1]:  # Exclude the target column
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]
    return df_cleaned

# Visualize the original data
plt.figure(figsize=(15, 10))
plt.title("Box Plot of Features Before Outlier Removal")
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(5, 6, i + 1)
    sns.boxplot(y=df[column])
    plt.title(column)
plt.tight_layout()
plt.show()

# Remove outliers from the DataFrame
df_no_outliers = remove_outliers_iqr(df)

# Visualize the cleaned data
plt.figure(figsize=(15, 10))
plt.title("Box Plot of Features After Outlier Removal")
for i, column in enumerate(df_no_outliers.columns[:-1]):
    plt.subplot(5, 6, i + 1)
    sns.boxplot(y=df_no_outliers[column])
    plt.title(column)
plt.tight_layout()
plt.show()

# Display the shape of the original and cleaned DataFrame
print("Original shape:", df.shape)
print("Shape after removing outliers:", df_no_outliers.shape)
```

Step: Import the dataset using libraries like Scikit-Learn, Pandas, or NumPy.

Step: Address any missing values by either removing them or imputing them with mean, median, or mode values.
 Missing data can lead to biased or inaccurate models. Handling missing values ensures that all data points contribute to the analysis.

Step: Use StandardScaler to standardize features, ensuring each has a mean of 0 and a standard deviation of 1.

6. Splitting the Dataset
Step: Divide the dataset into training and testing sets, often using an 80/20 or 70/30 split.
 Splitting the dataset allows for validation of the model on unseen data, helping to assess its performance and generalizability.
7. Feature Selection

# Logistic Regression
Logistic Regression is a statistical method used for binary classification tasks, predicting the likelihood of an outcome based on input features. It works by calculating a weighted sum of the features and applying a sigmoid function to convert this sum into a probability score between 0 and 1. For the breast cancer dataset, logistic regression is suitable because it can effectively classify tumors as benign or malignant, providing interpretable results and probabilities that help assess risk levels, making it valuable in medical diagnostics.


```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

# 2. Decision Tree Classifier
The Decision Tree Classifier is a machine learning algorithm that makes decisions based on a series of rules inferred from the data features. It works by recursively splitting the dataset into subsets based on the value of the features, creating a tree-like structure where each internal node represents a feature, each branch represents a decision rule, and each leaf node represents an outcome (e.g., benign or malignant). For the breast cancer dataset, the Decision Tree Classifier is suitable because it can handle both numerical and categorical data, provides clear visualizations of the decision-making process, and captures complex relationships in the data, making it easier to interpret and understand the factors influencing tumor classification.

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

3. Random Forest Classifier
The Random Forest Classifier is an ensemble learning method that builds multiple decision trees during training and merges their outputs for more accurate predictions. It works by creating a "forest" of decision trees, each trained on a random subset of the data and features. When making predictions, it aggregates the votes from all the trees to decide the final classification. For the breast cancer dataset, the Random Forest Classifier is suitable because it can effectively handle a large number of features, reduces the risk of overfitting compared to a single decision tree, and improves classification accuracy, making it robust for distinguishing between benign and malignant tumors.








```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

4. Support Vector Machine (SVM)

Support Vector Machine (SVM) is a supervised learning algorithm used for classification tasks that works by finding the optimal hyperplane that separates data points of different classes in a high-dimensional space. It identifies the support vectors, which are the data points closest to the hyperplane, and maximizes the margin between these support vectors and the hyperplane. For the breast cancer dataset, SVM is suitable because it can effectively classify complex, non-linear relationships between features, handles high-dimensional data well, and provides robust performance even when the classes are not perfectly separable, making it valuable for accurately distinguishing between benign and malignant tumors.

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(X_train, y_train)
```

3. Model Comparison

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize lists to store the results
results = []

# List of models and their names
models = [lr_model, dt_model, rf_model, svm_model, knn_model]
model_names = ["Logistic Regression", "Decision Tree", "Random Forest", "SVM", "k-NN"]

# Evaluate performance
for model, name in zip(models, model_names):
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    results.append((name, acc, prec, rec, f1))
    print(f"{name}: Accuracy={acc:.4f}, Precision={prec:.4f}, Recall={rec:.4f}, F1-score={f1:.4f}")

# Print all results in tabular form
results_df = pd.DataFrame(results, columns=["Model", "Accuracy", "Precision", "Recall", "F1 Score"])
print("\nOverall Performance:")
print(results_df)
```

```python
from sklearn.metrics import classification_report

# Evaluate performance for each model
for model, name in zip(models, model_names):
    y_pred = model.predict(x_test)
    report = classification_report(y_test, y_pred, target_names=["Benign", "Malignant"])
    print(f"{name}:\n{report}\n")
```
